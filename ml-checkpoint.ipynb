# chi-squared   χ^2 test 

Pearson's chi-squared test is used to determine whether there is a statistically significant difference (i.e., a magnitude of difference that is unlikely to be due to chance alone) between the expected frequencies and the observed frequencies in one or more categories of a so-called contingency table.

https://en.wikipedia.org/wiki/Chi-squared_test

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chisquare.html





```python
from scipy.stats import  chi2_contingency,chi2

# table form :
#                           
#  table=   [[         subCategory1  subCategory2 ...
#        Test-category1: 
#        Test-category2:
#                                                    ]]

table = [[10, 20, 30],
        [6,  9,  17]]


prob=0.95

# dof = deegrees of freedom
stat, p, dof, expected = chi2_contingency(table)

critical = chi2.ppf(prob, dof)

print(f'dof={dof}')
print(expected)
print(f'pVal={p}')
critical,stat
```

```python
table=[
    [100,1000],
    [70,1000]
]
stat, p, dof, expected = chi2_contingency(table)


print(p,f'\n \n{expected}')
```

```python
table=[
    [30,10],
    [15,25],
    [15,5]
]
stat, p, dof, expected = chi2_contingency(table)


print(p,f'\n \n{expected}')
```

# z-test
 When you can run a Z Test.
 Several different types of tests are used in statistics (i.e. f test, chi square test, t test). You would use a Z test if:
* Your sample size is greater than 30. Otherwise, use a t test.
* Data points should be independent from each other. In other words, one data point isn’t related or doesn’t affect another data point.
* Your data should be normally distributed. However, for large sample sizes (over 30) this doesn’t always matter.
* Your data should be randomly selected from a population, where each item has an equal chance of being selected.
* Sample sizes should be equal if at all possible.

https://en.wikipedia.org/wiki/Z-test

In a z-test, the sample is assumed to be normally distributed. A z-score is calculated with population parameters such as “population mean” and “population standard deviation” and is used to validate a hypothesis that the sample drawn belongs to the same population.

* The term "Z-test" is often used to refer specifically to the one-sample location test comparing the mean of a set of measurements to a given constant when the sample variance is known. For example, if the observed data X1, ..., Xn are (i) independent, (ii) have a common mean μ, and (iii) have a common variance σ2
The null hypothesis is that the mean value of X is a given number μ0. We can use X  as a test-statistic, rejecting the null hypothesis if X − μ0 is large.


* If the sample size is moderate or large, we can substitute the sample variance for σ2, giving a plug-in test. The resulting test will not be an exact Z-test since the uncertainty in the sample variance is not accounted for—however, it will be a good approximation unless the sample size is small.




```python
import numpy as np
from statsmodels.stats import weightstats as stests

l=[np.random.normal(100,12) for i in range(55)] 
testResult=stests.ztest(l,value=100)
pValue=testResult[1]
# if pValue is lower than a set threshold (most likely <0.1 or <0.05 we reject the null hypothesis )
print(f"p-value is: "+str(pValue),f'l mean is:{np.mean(l)}') 
```

# 2-sample Z test:
In two sample z-test   we are checking two independent data groups and deciding whether sample mean of two group is equal or not.

H0 : mean of two group is 0

H1 : mean of two group is not 0

```python
import numpy as np
from statsmodels.stats import weightstats as stests
counter=0
for i in range(100):
    sample1=[np.random.normal(100,10) for i in range(200)]
    sample2=[np.random.normal(100,10) for i in range(200)]
    ztest ,pval = stests.ztest(x1=sample1, x2=sample2,alternative='two-sided')
    if  pval<0.1:
        counter+=1
    
        print("p-value",pval)
print('counter:',counter)
counter/100

```

# T-test (2-sampled) 

(<30 data points, σ unknown)

The Independent Samples t Test or 2-sample t-test, compares the means of 2 independent samples in order to determine whether there is statistical evidence that the associated population means are <font color=blue> _significantly different_ .</font> The Independent Samples t Test is a parametric test. Like a z-test, a t-test also assumes a normal distribution of the sample. 



```python
from scipy.stats  import  ttest_ind 
import numpy as np

counter=0
for i in range(100):
    sample1=[np.random.normal(100,10) for i in range(20)]
    sample2=[np.random.normal(100,10) for i in range(20)]
    ttest,pval = ttest_ind(sample1,sample2)
    if  pval<0.1:
        counter+=1
    
        print("p-value",pval)
print('counter:',counter)
counter/100
```

# ANOVA (F-TEST) :

The z-test works well when dealing with two groups, but sometimes we want to compare more than two groups at the same time. For example, if we wanted to test whether voter age differs based on some categorical variable like race, we have to compare the means of each level or group the variable. We could carry out a separate z-test for each pair of groups, but when you conduct many tests you increase the chances of false positives. The analysis of variance or ANOVA is a statistical inference test that <font color=red>lets you compare multiple groups at the same time.</font>

## One Way F-test(Anova) :

It tell whether two or more groups are similar or not based on their mean similarity and f-score.

### What are the assumptions of a One-Way ANOVA?

Normality That each sample is taken from a normally distributed population


Sample independence – that each sample has been drawn independently of the other samples


Variance Equality – That the variance of data in the different groups should be the same

```python
from scipy import stats
counter=0
for i in range(100):
    sample1=[np.random.normal(100,10) for i in range(20)]
    sample2=[np.random.normal(100,10) for i in range(20)]
    sample3=[np.random.normal(100,10) for i in range(20)]

    F, pval = stats.f_oneway(sample1,sample2,sample3 )
    if  pval<0.1:
        counter+=1

        print("p-value",pval)
print('counter:',counter)
counter/100
```
